# Emplyment by Industry

```python
emp_ind = employ_indus.copy()
```

```python
emp_ind = emp_ind[emp_ind['Year']==2019]
```

```python
emp_ind = emp_ind.sort_values(by='Workforce by Industry and Gender', ascending=False)
```

```python
emp_ind.columns
```

```python
fig, ax = plt.subplots(1, figsize = (20,14))
squarify.plot(sizes = emp_ind['Workforce by Industry and Gender'], 
              label = emp_ind['Industry'][:11], 
              alpha=.7,
             pad = True)
plt.axis('off')
plt.title('Ames Workforce by Industry')
plt.show()
```

```python
px.histogram(x = emp_ind['Industry'][:10], y = emp_ind['Workforce by Industry and Gender'][:10])
```
